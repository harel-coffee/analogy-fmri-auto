```python
models = compile_models(write=False)
modelnames = ["rel", "humanratings", "w2vdiff", "rstpostprob9", "rstpostprob79", "numchar", "concatword", "typicality"]
# semantics_model = rsa.squareform(pu.load_mat_data(
#     os.path.join("analogy", "labels", "semdistmat144.mat"))["semdistmat"], checks=False)

def reduce_by_factor(rdm, factor):
    if rdm.ndim != 2:
        rdm = rsa.squareform(rdm)
    if (len(rdm) % factor) != 0:
        print("Factor issue, are you sure that's right?")
    return sum([rdm[i::factor, i::factor] for i in range(factor)])/factor

def downsample_rdms_by_factor(all_rdms, factor):
    downsampled_rdms = {}
    
    for mask, rdms in all_rdms.items():
        these_avg_rdms = []
        for rdm in rdms:
            these_rdms = []
            reduced_rdm = reduce_by_factor(rdm, factor)
            these_avg_rdms.append(reduced_rdm)
        downsampled_rdms[mask] = np.stack(these_avg_rdms)
    return downsampled_rdms
    

def create_models(models, modelnames, order):
    stacked_models_full = []
    stacked_models_avg = []
    stacked_models_subrel = []
    stacked_models_9rel = []
    for i, model in enumerate(modelnames):
        if (model == "numchar"):
            stacked_models_avg.append(
                rsa.rdm(np.array([
                    len(v) for v in order.ABTag.values[::2]
                ])[:, np.newaxis]))
            stacked_models_full.append(
                rsa.rdm(np.array([
                    len(v) for v in order.ABTag.values
                ])[:, np.newaxis]))
            stacked_models_subrel.append(
                rsa.squareform(
                    reduce_by_factor(stacked_models_avg[i], 2), 
                    checks=False))
            stacked_models_9rel.append(
                rsa.squareform(
                    reduce_by_factor(stacked_models_avg[i], 16), 
                    checks=False))
        elif (model == "typicality"):
            stacked_models_avg.append(
                rsa.rdm(np.array(
                    [models[model][v] for v in order.ABTag.values[::2]])))
            stacked_models_full.append(
                rsa.rdm(np.array(
                    [models[model][v] for v in order.ABTag.values])))
            stacked_models_subrel.append(
                rsa.squareform(
                    reduce_by_factor(stacked_models_avg[i], 2), 
                    checks=False))
            stacked_models_9rel.append(
                rsa.squareform(
                    reduce_by_factor(stacked_models_avg[i], 16), 
                    checks=False))
        else:
            stacked_models_avg.append(
                rsa.rdm(np.array(
                    [models[model][v] for v in order.ABTag.values[::2]]), 
                      metric="cosine"))
            stacked_models_full.append(
                rsa.rdm(np.array(
                    [models[model][v] for v in order.ABTag.values]), 
              metric="cosine"))
            stacked_models_subrel.append(
                rsa.squareform(
                    reduce_by_factor(stacked_models_avg[i], 2), 
                    checks=False))
            stacked_models_9rel.append(
                rsa.squareform(
                    reduce_by_factor(stacked_models_avg[i], 16), 
                    checks=False))
    #     stacked_models.append(semantics_model)

    return {"stacked_models": {"all": np.array(stacked_models_full), 
            "avg": np.array(stacked_models_avg),
            "subrel": np.array(stacked_models_subrel),
            "9rel": np.array(stacked_models_9rel)}, 
            "names": modelnames}


def models_to_df(models):
    models_df_dict={}
    for k, n in {"avg": "avg", "subrel": "subrel", "9rel": "downsampled"}.items():
        arr = pd.DataFrame(models["stacked_models"][k])
        names = [n for _ in range(len(arr))]
        models_df_dict[k] = pd.concat([
            pd.DataFrame({"type": names, "name": models["names"]}),
            arr
        ], axis=1)
    return models_df_dict


models_dict = create_models(models, modelnames, order)
models_df_dict = models_to_df(models_dict)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
def compile_models(write=False):
    # return dictionary of dictionaries
    w2vdiffs = pu.load_labels(os.path.join("analogy", "labels", "word2vec_diffs.csv"))
    humanratings = pu.load_labels(os.path.join("analogy", "labels", "humanratings.csv"), skiprows=2)
    rstpostprob9 = pu.load_mat_data(os.path.join("analogy", "labels", "rstpostprob9.mat"))
    rstpostprob79 = pu.load_mat_data(os.path.join("analogy", "labels", "rstpostprob79.mat"))
    concatword = pu.load_mat_data(os.path.join("analogy", "labels", "w2vconcat.mat"))
    mat_concatword = {}
    mat_rstpostprob79 = {}
    mat_humanratings = {}
    mat_rstpostprob9 = {}
    mat_mainrel = {}
    mat_subrel = {}
    mat_rel = {}
    mat_sem = {}
    mat_intuit = {}
    mat_w2v = {}
    wordpairs = []
    mainenc = OneHotEncoder(4)
    subenc = OneHotEncoder(10)
    for i in range(len(rstpostprob79["wordpair"])):
        wordpair = rstpostprob79["wordpair"][i, 0][0]
        wordpairs.append(wordpair)
        ci = np.where(concatword["wordpair"][:, :] == wordpair)[0][0]
        mat_concatword[wordpair] = concatword["concwordmat"].astype(np.float)[ci]
        mat_rstpostprob9[wordpair] = rstpostprob9["rstpostprob_sm"].astype(np.float)[i]
        mat_rstpostprob79[wordpair] = rstpostprob79["rstpostprob"].astype(np.float)[i]
        mat_humanratings[wordpair] = humanratings[humanratings.wordpair==wordpair].as_matrix()[0, 1:].astype(np.float)
        mat_w2v[wordpair] = w2vdiffs[w2vdiffs.wordpair==wordpair].as_matrix()[0, 1:].astype(np.float)
        
        mat_mainrel[wordpair] = mainenc.fit_transform(rstpostprob79["wordpair"][i, 1] - 1).toarray()[0, :-1]
        mat_subrel[wordpair] = subenc.fit_transform(rstpostprob79["wordpair"][i, 2] - 1).toarray()[0, :-1]
        mat_rel[wordpair] = np.hstack([mat_mainrel[wordpair], mat_subrel[wordpair]])
        mat_sem[wordpair] = np.hstack([mat_mainrel[wordpair][0] + mat_mainrel[wordpair][1], 
                                       mat_mainrel[wordpair][2]])
        mat_intuit[wordpair] = np.hstack([mat_sem[wordpair], mat_rel[wordpair]])
    models = {
          "wordpairs": wordpairs, 
          "humanratings": mat_humanratings,
          "rstpostprob79": mat_rstpostprob79,
          "rstpostprob9": mat_rstpostprob9,
          "w2vdiff": mat_w2v,
          "mainrel": mat_mainrel,
          "subrel": mat_subrel,
          "rel": mat_rel,
          "sem": mat_sem,
          "intuit": mat_intuit,
          "concatword": mat_concatword
             }
    return models

#     labels_df = pd.DataFrame(labels).drop_duplicates(subset="wordpair").reset_index(drop=True) # pd.DataFrame.from_dict(labels, columns=column_names)
#     if write:
#         labels_df.to_csv(os.path.join("analogy", "labels", "pd_allmodels.csv"), index=False)
#     return labels_df
                            
    
models = compile_models(write=False)

# order the trials the same for all subjects so the RDMs can be stacked
# order = {}
# order["full"] = pu.load_labels("analogy/labels/trialorder_rsa_absorted.csv")
# order["match"] = pu.load_labels("analogy/labels/trialorder_rsa_match.csv")
# order["nomatch"] = pu.load_labels("analogy/labels/trialorder_rsa_nomatch.csv")

order = pu.load_labels("analogy/labels/trialorder_rsa_absorted.csv")

def reorder_data_pymvpa(data, old_order, new_order):
    return np.vstack([data[old_order.trialtag == v] for v in new_order.TrialTag])

def reorder_data_lss(data, old_order, new_order):
    return np.vstack([data[old_order.TrialTag == v] for v in new_order.TrialTag])


#pymvpa utils

def load_data_pymvpa(sub, maskname, normalize=True):
    imgFile = os.path.join(paths['root'], 'derivatives', sub, 'betas', pu.format_bids_name(sub, 'task-analogy', 'betas-pymvpa.nii.gz'))
    mask = pu.load_img(paths['root'], 
                      'derivatives', sub, 'masks', 
                      maskname + '.nii.gz')
    labels = pu.load_labels(paths['root'], 
                           'derivatives', sub, 'betas', 
                           pu.format_bids_name(sub, 'task-analogy', 'events-pymvpa.tsv'), 
                           sep='\t', index_col=0)
    maskedImg = pa.mask_img(imgFile, mask)
    conditionSelector = np.where(labels['ab'] == 1)
    
    if normalize:
        resids = []
        # whiten the data
        for r in range(8):
            residFile = os.path.join(paths['root'], 'derivatives', sub, 'func', 
                                   pu.format_bids_name(sub, 'task-analogy', 
                                                       "run-0{}".format(r+1), 
                                                       'resids-pymvpa.nii.gz'))
            resids.append(pa.mask_img(residFile, mask))


        fmri_data = rsa.noise_normalize_beta(maskedImg[conditionSelector], np.vstack(resids))
    else:
        fmri_data = maskedImg[conditionSelector]
        
    these_labels = labels.iloc[conditionSelector]
    bg_image = pu.load_img(os.path.join(paths['root'], 
                                        'derivatives', 
                                        sub, 'reg', 'BOLD_template.nii.gz'))
    return fmri_data, these_labels, bg_image, mask


# LSS utils
def load_data_lss(sub, maskname, t="tstat", normalize=True):
#     imgFile = os.path.join(paths['root'], 'derivatives', sub, 'betas', pu.format_bids_name(sub, 'task-analogy', 'betas-pymvpa.nii.gz'))
    mask = pu.load_img(paths['root'], 
                      'derivatives', sub, 'masks', 
                      maskname + '.nii.gz')
#     labels = pu.load_labels(paths['root'], 
#                            'derivatives', sub, 'betas', 
#                            pu.format_bids_name(sub, 'task-analogy', 'events-pymvpa.tsv'), 
#                            sep='\t', index_col=0)
#     maskedImg = pa.mask_img(imgFile, mask)
    
#     resids = []
    labels = []
    imgs = []
    # whiten the data
    for r in range(8):
        imgFile = os.path.join(paths['root'], 'derivatives', sub, 'betas', 
                               "{}_task-analogy_run-0{}_betas-{}-LSS.nii.gz".format(sub, r+1, t))
        labels.append(pu.load_labels(paths["root"],
                           'derivatives', sub, 'func', 
                            "{}_task-analogy_run-0{}_events.tsv".format(sub, r+1),
                           sep='\t'))
        residFile = os.path.join(paths['root'], 'derivatives', sub, 'func', 
                               pu.format_bids_name(sub, 'task-analogy', 
                                                   "run-0{}".format(r+1), 
                                                   'resids-lss.nii.gz'))
        if normalize:
            maskedImg = pa.mask_img(imgFile, mask)
            resids = pa.mask_img(residFile, mask)
            imgs.append(rsa.noise_normalize_beta(maskedImg, resids))
        else: 
            imgs.append(pa.mask_img(imgFile, mask))

#         resids.append(pa.mask_img(residFile, mask))
#         run_data = noise_normalize_beta()
    labels = pd.concat(labels).reset_index()
    conditionSelector = np.where(labels['AB'] == 1)
    fmri_data = np.vstack(imgs)
    
#     fmri_data = rsa.noise_normalize_beta(maskedImg[conditionSelector], np.vstack(resids))

    these_labels = labels.iloc[conditionSelector]
    fmri_data = fmri_data[conditionSelector]
    bg_image = pu.load_img(os.path.join(paths['root'], 
                                        'derivatives', 
                                        sub, 'reg', 'BOLD_template.nii.gz'))
    return fmri_data, these_labels, bg_image, mask





def generate_rdms(method, maskname, order, 
                  loadfunc, loadfunc_args=None, 
                  reorderfunc=reorder_data_lss):
    full = []
#     match = []
#     nomatch = []
    for sub in projectSettings["subjects"].keys():
        fmri_data, these_labels, _, _ = loadfunc(sub, maskname, **loadfunc_args)
        fmri_reordered_data = reorderfunc(fmri_data, these_labels, order)
        if method == "avg":
            full.append(pa.rdm((fmri_reordered_data[::2] + fmri_reordered_data[1::2])/2, metric="euclidean"))
        else:
            full.append(pa.rdm(fmri_reordered_data, metric="euclidean"))
#         match.append(pa.rdm(reorderfunc(fmri_data, these_labels, order["match"]), metric="euclidean"))
#         nomatch.append(pa.rdm(reorderfunc(fmri_data, these_labels, order["nomatch"]), metric="euclidean"))
    return np.array(full)  #, np.array(match), np.array(nomatch)


def all_subject_rdms_pymvpa(maskname, order, m="avg", save=True, load=True):
    rdmfunc_args = {"loadfunc": load_data_pymvpa, 
                "loadfunc_args": {"normalize": True}, 
                "reorderfunc": reorder_data_pymvpa}
    if load:
        try:
            rdms = np.load(os.path.join(paths['cloud'], 
                                     "analysis", 
                                     "rsa", "rdms", "{}_trials".format(m),
                                     "subject-rdms_{}-ab-pymvpa-whitened.npy".format(maskname)))

            save = False
        except FileNotFoundError:
            rdms = generate_rdms(m, maskname, order, **rdmfunc_args)
    else:
        rdms = generate_rdms(m, maskname, order, **rdmfunc_args)
    
    if save:
        np.save(os.path.join(paths['root'], 
                             "analysis", 
                             "rsa", "rdms", "{}_trials".format(m),
                             "subject-rdms_{}-ab-pymvpa-whitened.npy".format(maskname)), 
                rdms)
    return rdms


def all_subject_rdms_lss(maskname, order, m="avg", save=True, load=True):
    rdmfunc_args = {"loadfunc": load_data_lss, 
                "loadfunc_args": {"t": "tstat", "normalize": True}, 
                "reorderfunc": reorder_data_lss}
    if load:
        try:
            rdms = np.load(os.path.join(paths['cloud'], 
                                     "analysis", 
                                     "rsa", "rdms", "{}_trials".format(m),
                                     "subject-rdms_{}-ab-tstat-whitened.npy".format(maskname)))
            save = False
        except FileNotFoundError:
            rdms = generate_rdms(m, maskname, order, **rdmfunc_args)
    else:
        rdms = generate_rdms(m, maskname, order, **rdmfunc_args)
    
    if save:
        np.save(os.path.join(paths['root'], 
                             "analysis", 
                             "rsa", "rdms", "{}_trials".format(m),
                             "subject-rdms_{}-ab-tstat-whitened.npy".format(maskname)), 
                rdms)
    return rdms

# def all_subject_rdms_pymvpa(maskname, order, m="avg", save=True, load=True):
#     rdmfunc_args = {"loadfunc": load_data_pymvpa, 
#                 "loadfunc_args": {"normalize": False}, 
#                 "reorderfunc": reorder_data_pymvpa}
#     if load:
#         try:
#             rdms = np.load(os.path.join(paths['cloud'], 
#                                      "analysis", 
#                                      "rsa", "rdms", "{}_trials".format(m),
#                                      "subject-rdms_{}-ab-pymvpa.npy".format(maskname)))

#             save = False
#         except FileNotFoundError:
#             rdms = generate_rdms(m, maskname, order, **rdmfunc_args)
#     else:
#         rdms = generate_rdms(m, maskname, order, **rdmfunc_args)
    
#     if save:
#         np.save(os.path.join(paths['root'], 
#                              "analysis", 
#                              "rsa", "rdms", "{}_trials".format(m),
#                              "subject-rdms_{}-ab-pymvpa.npy".format(maskname)), 
#                 rdms)
#     return rdms


# def all_subject_rdms_lss(maskname, order, m="avg", save=True, load=True):
#     rdmfunc_args = {"loadfunc": load_data_lss, 
#                 "loadfunc_args": {"t": "tstat", "normalize": False}, 
#                 "reorderfunc": reorder_data_lss}
#     if load:
#         try:
#             rdms = np.load(os.path.join(paths['cloud'], 
#                                      "analysis", 
#                                      "rsa", "rdms", "{}_trials".format(m),
#                                      "subject-rdms_{}-ab-tstat.npy".format(maskname)))
#             save = False
#         except FileNotFoundError:
#             rdms = generate_rdms(m, maskname, order, **rdmfunc_args)
#     else:
#         rdms = generate_rdms(m, maskname, order, **rdmfunc_args)
    
#     if save:
#         np.save(os.path.join(paths['root'], 
#                              "analysis", 
#                              "rsa", "rdms", "{}_trials".format(m),
#                              "subject-rdms_{}-ab-tstat.npy".format(maskname)), 
#                 rdms)
#     return rdms

modelnames = ["rel", "humanratings", "w2vdiff", "rstpostprob9", "rstpostprob79", "numchar", "concatword"]
# semantics_model = rsa.squareform(pu.load_mat_data(
#     os.path.join("analogy", "labels", "semdistmat144.mat"))["semdistmat"], checks=False)

def create_models(models, modelnames, order):
    stacked_models_full = []
    stacked_models_avg = []
    for model in modelnames:
        if model == "numchar":
            stacked_models_avg.append(rsa.rdm(np.array([len(v) for v in order.ABTag.values[::2]])[:, np.newaxis]))
            stacked_models_full.append(rsa.rdm(np.array([len(v) for v in order.ABTag.values])[:, np.newaxis]))

        else:
            stacked_models_avg.append(rsa.rdm(np.array([models[model][v] for v in order.ABTag.values[::2]]), 
                      metric="cosine"))
            stacked_models_full.append(rsa.rdm(np.array([models[model][v] for v in order.ABTag.values]), 
              metric="cosine"))
    #     stacked_models.append(semantics_model)

    return {"all": np.array(stacked_models_full), 
            "avg": np.array(stacked_models_avg)}

stacked_models = create_models(models, modelnames, order)
```

```python
modelnames = ["rel", "humanratings", "w2vdiff", "rstpostprob9", "rstpostprob79", "numchar", "concatword"]
# semantics_model = rsa.squareform(pu.load_mat_data(
#     os.path.join("analogy", "labels", "semdistmat144.mat"))["semdistmat"], checks=False)
        
    
def run_rsa(subject_rdms, stacked_models, d="full"):
    res = {}
    for mask, rdms in subject_rdms.items():
        res[mask] = {}
        res[mask]["corr"] = np.array([rsa.rdm(np.vstack([rdm, stacked_models]), metric="spearman")[0] for rdm in rdms])
        res[mask]["pval"] = np.array([rsa.rdm(np.vstack([rdm, stacked_models]), metric="spearman")[1] for rdm in rdms])
    return res
 
```

```python
plot_modelnames = ["rel", "humanratings", "w2vdiff", "rstpostprob9", "rstpostprob79", "numchar", "concatword"]
def plot_results(corrs, stats, roi, d="pymvpa", ceilings=None, save=False):
    print(stats[d][roi])
    means = (1-np.array(corrs[d][roi]["corr"])[:,0:len(plot_modelnames)]).mean(0)
    stds = (1-np.array(corrs[d][roi]["corr"])[:,0:len(plot_modelnames)]).std(0)/np.sqrt(15)
    # modelnames = ["subrel", "mainrel", "sem", "rel", "intuit", "typicality", "rst9", "rst79"]
    error_config = {'ecolor': '0.3'}

    rects1 = plt.bar(plot_modelnames, means,
                     color='b',
                     yerr=stds,
                     error_kw=error_config,
                     label='Model fits')
    plt.ylim([-.1, .3])
    if ceilings is not None:
        upper = ceilings[roi].mean(1)
        lower = ceilings[roi].mean(1)
        plt.plot([0, len(plot_modelnames)], [lower, lower], 'r-', 
                 linewidth=.5, label="lower")
        plt.plot([0, len(plot_modelnames)], [upper, upper], 'r-', 
                 linewidth=.5, label="upper")
        
    plt.title("{}-{}".format(roi, d))
    if save:
        plt.savefig("/home/njchiang/data/CloudStation/Grad/Research/Analogy/Figures/{}-{}-modelcomp.png".format(roi, d))
    return
```

```python
# Downsample ROIs:
from scipy.spatial.distance import squareform
def downsample_rdms(all_rdms, order):
    avg_rdms = {}
    n_per_cond = int(len(order) / len(order.unique()))
    indices = np.arange(len(order.unique()) + 1) * n_per_cond
    for mask, rdms in all_rdms.items():
        these_avg_rdms = []
        for rdm in rdms:
            this_full_rdm = squareform(rdm)
            this_avg_rdm = np.zeros([len(order.unique()), len(order.unique())])
            for i in range(len(indices) - 1):
                for j in range(len(indices) - 1):
                    this_avg_rdm[i,j] = this_full_rdm[indices[i]:indices[i+1], 
                                                      indices[j]:indices[j+1]].mean()
                
            these_avg_rdms.append(this_avg_rdm)
            avg_rdms[mask] = np.stack(these_avg_rdms)
    return avg_rdms
    
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
def compile_models(write=False):
    # return dictionary of dictionaries
    w2vdiffs = pu.load_labels(os.path.join("analogy", "labels", "word2vec_diffs.csv"))
    humanratings = pu.load_labels(os.path.join("analogy", "labels", "humanratings.csv"), skiprows=2)
    rstpostprob9 = pu.load_mat_data(os.path.join("analogy", "labels", "rstpostprob9.mat"))
    rstpostprob79 = pu.load_mat_data(os.path.join("analogy", "labels", "rstpostprob79.mat"))
    mat_rstpostprob79 = {}
    mat_humanratings = {}
    mat_rstpostprob9 = {}
    mat_mainrel = {}
    mat_subrel = {}
    mat_rel = {}
    mat_sem = {}
    mat_intuit = {}
    mat_w2v = {}
    wordpairs = []
    mainenc = OneHotEncoder(4)
    subenc = OneHotEncoder(10)
    for i in range(len(rstpostprob79["wordpair"])):
        wordpair = rstpostprob79["wordpair"][i, 0][0]
        wordpairs.append(wordpair)
        mat_rstpostprob9[wordpair] = rstpostprob9["rstpostprob_sm"].astype(np.float)[i]
        mat_rstpostprob79[wordpair] = rstpostprob79["rstpostprob"].astype(np.float)[i]
        mat_humanratings[wordpair] = humanratings[humanratings.wordpair==wordpair].as_matrix()[0, 1:].astype(np.float)
        mat_w2v[wordpair] = w2vdiffs[w2vdiffs.wordpair==wordpair].as_matrix()[0, 1:].astype(np.float)
        
        mat_mainrel[wordpair] = mainenc.fit_transform(rstpostprob79["wordpair"][i, 1] - 1).toarray()[0, :-1]
        mat_subrel[wordpair] = subenc.fit_transform(rstpostprob79["wordpair"][i, 2] - 1).toarray()[0, :-1]
        mat_rel[wordpair] = np.hstack([mat_mainrel[wordpair], mat_subrel[wordpair]])
        mat_sem[wordpair] = np.hstack([mat_mainrel[wordpair][0] + mat_mainrel[wordpair][1], 
                                       mat_mainrel[wordpair][2]])
        mat_intuit[wordpair] = np.hstack([mat_sem[wordpair], mat_rel[wordpair]])
    models = {
          "wordpairs": wordpairs, 
          "humanratings": mat_humanratings,
          "rstpostprob79": mat_rstpostprob79,
          "rstpostprob9": mat_rstpostprob9,
          "w2vdiff": mat_w2v,
          "mainrel": mat_mainrel,
          "subrel": mat_subrel,
          "rel": mat_rel,
          "sem": mat_sem,
          "intuit": mat_intuit
             }
    return models
#         labels.append({"wordpair": rstpostprob79["wordpair"][i, 0][0],
#                       "mainrel": rstpostprob79["wordpair"][i, 1][0, 0],
#                       "subrel": rstpostprob79["wordpair"][i, 2][0, 0],
#                       "humanratings": humanratings[humanratings.wordpair==wordpair].as_matrix()[0, 1:].astype(np.float),
#                       "rstpostprob9": rstpostprob9["rstpostprob_sm"].astype(np.float)[i],
#                       "rstpostprob79": rstpostprob79["rstpostprob"].astype(np.float)[i]
#                       }
#                      )
#     labels_df = pd.DataFrame(labels).drop_duplicates(subset="wordpair").reset_index(drop=True) # pd.DataFrame.from_dict(labels, columns=column_names)
#     if write:
#         labels_df.to_csv(os.path.join("analogy", "labels", "pd_allmodels.csv"), index=False)
#     return labels_df
                            
```
